```python
indicies=pd.Series(join_data1.index,index=join_data1['Title']).drop_duplicates()
```

```python
indicies.head()
```

```python
def give_rec(title,sig=sig):
    idx=indicies[title]
    sig_score=list(enumerate(sig[idx]))
    sig_score=sorted(sig_score,key=lambda x1: x1[1], reverse=True)
    sig_score=sig_score[1:11]
    movie_indices=[i[0] for i in sig_score]
    
    return join_data1['Title'].iloc[movie_indices]
```
